```python
def process_result(marks_list):
    
    total = 0
    
    for marks in marks_list:
        total += marks
    
    average = total / len(marks_list)
    
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

marks = [60, 45, 70, 55, 40]

avg, result = process_result(marks)

print("Average Marks:", avg)
print("Result:", result)
```
