```python
def calculate_fare(distance, peak):
    
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)
    
    return fare

while True:
    
    distance = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak_hour)
    
    print("Total Cab Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ")
    
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator ")
        break
```
