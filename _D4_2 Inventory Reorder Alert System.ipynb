```python
# Dictionary of products and their stock quantities
inventory = {
    "canola": 15,
    "Sunflower": 20,
    "Oil": 8,
    "wheat": 50,
    "biscuit": 20
}

def check_stock(product, quantity):
    if quantity < 15:
        print(product, "→ Reorder Alert (Stock:", quantity, ")")
    else:
        print(product, "→ Stock OK (Stock:", quantity, ")")

for product, quantity in inventory.items():
    check_stock(product, quantity)
```
