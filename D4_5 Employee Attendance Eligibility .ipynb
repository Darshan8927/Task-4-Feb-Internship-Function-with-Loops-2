```python
def check_eligibility(attendance_list):
    
    total_days = len(attendance_list)
    present_days = 0
    
    for status in attendance_list:
        if status == "P":
            present_days += 1
    
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


attendance= []
days=int(input("Enter todays of Working"))
for i in range(days):
    status = input(f"Enter attendance for Day {i+1} (P/A): ")
    attendance.append(status)

percent, result = check_eligibility(attendance)

print("\nAttendance Percentage:", percent, "%")
print("Eligibility Status:", result)
```
